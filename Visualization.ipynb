# Learnings for Visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Set basic parameters for plot

## Size of plot

```python
plt.figure(figsize=(5,5))
```

# Types

## Line graph

```python
plt.plot([1,2,3,4],[1,2,3,4])
```

## Dots

```python
plt.plot([1,2,3,4],[1,2,3,4], "bo")
```

## Bar Chart (stacked)

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.bar(stacked=True)
```

# Labeling, Formating and Commenting

```python
sns.scatterplot([1,2,3,4],[1,2,3,4])
plt.text(3.2,3, "Test-Punkt")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Title")
plt.ylim(0,5)
plt.xlim(0,5);
```

```python
sns.scatterplot([1,2,3,4],["Tick 1", "Tick 2", "Tick 3", "Tick 4"])
plt.text(3.2, 2, "Test-Punkt") #As Y-Labels are strings, the y-position of the text needs to reference \
                                    # the index position, instead of a value
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Title")

plt.xlim(0,5);
```

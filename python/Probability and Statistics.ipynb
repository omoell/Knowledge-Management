```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
from scipy.stats import binom
```

# Central Limit Theorem

Sample means drawn from the same population will over time adapt to a normal distribution. This effect normally gets transparent >30 samples.

```python
# Calcucalte 10, 100, 1000 sample means with sample size 30
means10 = [np.random.randint(1, 7, 30).mean() for i in range(10)]
means100 = [np.random.randint(1, 7, 30).mean() for i in range(100)]
means1000 = [np.random.randint(1, 7, 30).mean() for i in range(1000)]

# Create plot area with 3 plots in one row
fig = plt.figure(figsize=(20,5))
ax10 = fig.add_subplot(131)
ax100 = fig.add_subplot(132)
ax1000 = fig.add_subplot(133)

# Create histograms of the means
ax10.hist(means10)
ax10.set_title('10 means')
ax100.hist(means100)
ax100.set_title('100 means')
ax1000.hist(means1000)
ax1000.set_title('1,000 means')

# Show a histogram of the means
plt.show()
```

# Law of large numbers

Over time (repetition of the event under same conditions) the relative frequency will be close to the expected frequency (e.g. 1/6 for each number of a dice)

```python
# Roll the dice 30, 100 and 100,000 times
dice30 = [np.random.randint(1, 7, 30)]
dice100 = [np.random.randint(1, 7, 100)]
dice100000 = [np.random.randint(1, 7, 100000)]

# Create plot area with 3 plots in one row
fig = plt.figure(figsize=(20,5))
ax30 = fig.add_subplot(131)
ax100 = fig.add_subplot(132)
ax100000 = fig.add_subplot(133)

# Create histograms of the means
ax30.hist(dice30)
ax30.set_title('30 rolls')
ax100.hist(dice100)
ax100.set_title('100 rolls')
ax100000.hist(dice100000)
ax100000.set_title('100,000 rolls')

# Show histogram of results of dice rolls
plt.show()
```

# Probability distributions

Common distributions:



## Bernoulli Distribution

Distribution with two outcomes: 0 or 1, e.g. coin flip.

```python
# Generate bernoulli data for coin flips (fair coin)
data = bernoulli.rvs(p=0.5, size=1000)

# Plot distribution
plt.hist(data)
plt.show()
```

## Binomial Distribution

A count (number of times) something happened: Sum of the outcome of a bernoulli experiment: No. of Trials (n), No. of Success (k), Probability of Success (p).

Example:
Consider a game where you are trying to make a ball in a basket. You are given 10 shots and you know that you have an 80% chance of making a given shot. To simplify things, assume each shot is an independent event.

```python
# Generate binomial data
data = binom.rvs(n=10, p=0.8, size=1000)

# Plot the distribution
plt.hist(data)
plt.show()

# Assign and print probability of 8 or less successes
prob1 = binom.cdf(k=8, n=10, p=0.8)
print(prob1)

# Assign and print probability of all 10 successes
prob2 = binom.pmf(k=10, n=10, p=0.8)
print(prob2)
```
